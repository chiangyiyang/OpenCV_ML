## Reference

# Download a sample image

```python
import urllib.request

url = 'https://upload.wikimedia.org/wikipedia/commons/1/16/HDRI_Sample_Scene_Balls_%28JPEG-HDR%29.jpg'
urllib.request.urlretrieve(url, "sample.jpg")

sample_file = 'sample.jpg'
```

# Read an image

```python
import numpy as np
import cv2

# Load an color image in grayscale
img = cv2.imread(sample_file,0)
```

# Display an image

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL) #you can resize window
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Write an image

```python
cv2.imwrite('sample_gray.png',img)
```

# Using Matplotlib

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread(sample_file,0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
